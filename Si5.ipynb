**Ejercicio 5** 05 Auditoría de Desempeño para Negocios Digitales:

En este ejercicio vamos a analizar un dataset de transacciones y determinar si la estrategia de M-Commerce (App Móvil) es más rentable que el E-Commerce tradicional (Web Desktop).

Que canal genera mas ingresos?

```python
import pandas as pd
import numpy as np

# Semilla para reproducibilidad
np.random.seed(42)

# Crear datos simulados
data = {
    'ID_Transaccion': range(1001, 1051),
    'Canal': np.random.choice(['Web Desktop', 'App Móvil'], 50, p=[0.4, 0.6]),
    'Metodo_Pago': np.random.choice(['Tarjeta Crédito', 'PayPal', 'Transferencia'], 50),
    'Total_Venta': np.random.randint(20, 500, 50), # Ventas entre $20 y $500
    'Categoria': np.random.choice(['Electrónica', 'Moda', 'Hogar'], 50)
}

df = pd.DataFrame(data)

df
```

```python
#Kpis
#1. Volumen de ventas por canal.
#2. Ticket promedio (promedio de  ingresos).
#3. Preferencia de pagos (En cada comerce cual metodo de pago se uso mas).
#4. Categoria por canal (Que se compra mas en cada uno de los comerce).
#"2"
# Umbral de cuotas.
```

```python
df['Canal'].value_counts()
```

El canal más ocupado es el es M-commerce (App móvil)

```python
ingreso_totalcanal = df.groupby('Canal')['Total_Venta'].sum()
ingreso_totalcanal
```

```python
diferencia = ingreso_totalcanal['App Móvil'] - ingreso_totalcanal['Web Desktop']
diferencia
```

En M-commerce (App móvil) genero mas ingresos totales con una diferencia de 1230

```python
ingreso_total = df['Total_Venta'].sum()
ingreso_total
```

```python
ingreso_prom = df.groupby('Canal')['Total_Venta'].mean()
ingreso_prom
```
