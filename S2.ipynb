```python
import pandas
```

```python
filename = 'https://raw.githubusercontent.com/ivan-carrera/isi25b/refs/heads/main/data/supermarket.csv'
```

```python
df = pandas.read_csv(filename)
```

```python
print(df.head())
```

```python
print(df.iloc[100]['Fecha'])
```

```python
df_centro = df[df['Sucursal'] == 'Centro']
print("Total Centro:", df_centro['Valor_Total'].sum())
```

```python
df_norte = df[df['Sucursal'] == 'Norte']
print("Total Norte:", df_norte['Valor_Total'].sum())
```

```python
df_sur = df[df['Sucursal'] == 'Sur']
print("Total Sur:", df_sur['Valor_Total'].sum())
```
